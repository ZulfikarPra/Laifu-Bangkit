```python
url = '/content/Laifu_Food.csv'
df = pd.read_csv(url)
df.info()
```

```python
output = df[['Food Name', 'Ingredients']]

# Drop duplicates based on 'Food Name'
output = output.drop_duplicates(subset='Food Name')

# Add a new 'Index' column with sequential indices
output['Index'] = range(len(output))

output
```

```python
#Drop Food ID, Rating, Cuisine Column and Empty Column
df.drop(columns=['Food ID','Rating','Cuisine'], inplace=True)
df.dropna(inplace=True)

#Replace BMI Category into List of Number
df['BMI Category'].replace({'Underweight':0,'Normal':0,'Normal Weight':0,'Overweight':1,'Obese':2},inplace=True)

#Replace Spice Level into List of Number
unique_spice = df['Spice Level'].unique()
df['Spice Level'].replace(to_replace=unique_spice, value=list(range(len(unique_spice))), inplace=True)

#Replace Physical Activity Level into List of Number
unique_physical = df['Physical Activity Level'].unique()
df['Physical Activity Level'].replace(to_replace=unique_physical, value=list(range(len(unique_physical))), inplace=True)

#Replace Stress Level into List of Number
unique_stress = df['Stress Level'].unique()
df['Stress Level'].replace(to_replace=unique_stress, value=list(range(len(unique_stress))), inplace=True)

#Replace Preference into List of Number
unique_preference = df['Preference'].unique()
df['Preference'].replace(to_replace=unique_preference, value=list(range(len(unique_preference))), inplace=True)

#Replace Recommended For into List of Number
unique_reccomended = df['Recommended For'].unique()
df['Recommended For'].replace(to_replace=unique_reccomended, value=list(range(len(unique_reccomended))), inplace=True)

#Replace Food Name into List of Number
unique_food = df['Food Name'].unique()
df['Food Name'].replace(to_replace=unique_food, value=list(range(len(unique_food))), inplace=True)

#Split Blood Pressure Column into Sistole and Diastole
df[['Sistole', 'Diastole']] = df['Blood Pressure'].str.split('/', expand=True)
df.drop(columns=['Blood Pressure'], inplace=True)

df[['Sistole', 'Diastole', 'Recommended For', 'Preference']] = df[['Sistole', 'Diastole', 'Recommended For', 'Preference']].astype(np.int64)
```

```python
x = df.drop(columns=['Food Name', 'Ingredients'])
y = df['Food Name']
```

```python
x.dtypes
```

```python
train_size = int(0.7 * x.shape[0])

x_train = x[:train_size]
y_train = y[:train_size]

x_valid = x[train_size:]
y_valid = y[train_size:]
```

```python
tf_dataset = tf.data.Dataset.from_tensor_slices((x_train.values, y_train.values)).batch(32)
tf_dataset_val = tf.data.Dataset.from_tensor_slices((x_valid.values, y_valid.values)).batch(32)
```

```python
for features, targets in tf_dataset.take(1):
  print ('Features: {}, Target: {}'.format(features, targets))
```

```python
from keras.callbacks import ModelCheckpoint

model = tf.keras.models.Sequential([
    tf.keras.layers.Embedding(1000, 1024, input_length=8),
    tf.keras.layers.Conv1D(128, 5, activation='relu'),
    # tf.keras.layers.LSTM(256, return_sequences=True),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(54, activation='sigmoid')
])

model.summary()

checkpoint = ModelCheckpoint(filepath = './checkpoint//model.{epoch:02d}-{val_accuracy:06f}.h5', monitor = 'val_accuracy', save_best_only = True)

model.compile(loss='sparse_categorical_crossentropy',
                  optimizer = tf.keras.optimizers.Adam(learning_rate = 0.0005),
                  metrics=['accuracy'])

model.fit(tf_dataset, epochs = 100,  validation_data = tf_dataset_val, callbacks = [checkpoint])
```

```python
import numpy as np
new_x = np.array([1, 1, 0, 1, 1, 0, 140, 90])
new_x = {'Spice Level': new_x[0], 'Recommended For': new_x[1], 'Physical Activity Level': new_x[2], 'Stress Level': new_x[3],
         'BMI Category': new_x[4], 'Preference': new_x[5], 'Sistole': new_x[6], 'Diastole': new_x[7]}
new_x = pd.DataFrame(data=new_x, index=[0])
new_x
```

```python
new_x = tf.convert_to_tensor(new_x)
new_x
```

```python
result = model.predict(new_x)
result
```

```python
max_index = np.argmax(result)
selected_row = output.loc[max_index]
print('Food recommendation:', selected_row['Food Name'])
print('Ingredients:',selected_row['Ingredients'])
```
